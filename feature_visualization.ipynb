```python
import glob

# first_index crosspond to Dastgah number
# Second index crosspond to instrument number =

root_folder  = "ML_prj_2023"
dastgah_folder_names = [f"D_{i}" for i in range(7)]
instument_folder_names = [f"I_{i}" for i in range(6)]

file_names = []
file_nums = 0

for i, dastgah_folder_name in enumerate(dastgah_folder_names):

  dastgah_file_names = []

  for j , instument_folder_name in enumerate(instument_folder_names):
      
      file_names_regex = f"{root_folder}/{dastgah_folder_name}/{instument_folder_name}/*"

      # print(file_names_regex)

      instrument_file_names = glob.glob(file_names_regex)

      dastgah_file_names.append(instrument_file_names)

      file_nums += len(instrument_file_names)
  
  file_names.append(dastgah_file_names)
     

file_names, file_nums
```

```python
import numpy as np
import pickle

import librosa.display
import librosa
import matplotlib.pyplot as plt
```

```python
class_names = ["Shoor","Segah","Mahoor","Homayoun","Rast Panjgah","Nava","Chargah"]

selected_files = ["ML_prj_2023/D_0/I_0/Farzad_Soltani_tar_shoor.mp3",
                  "ML_prj_2023/D_1/I_1/Rohhab-Shekarchi.mp3",
                  "ML_prj_2023/D_2/I_4/Khavaran-Jahangiri.mp3",
                  "ML_prj_2023/D_3/I_2/santoor.mp3",
                  "ML_prj_2023/D_4/I_5/shahramnazeri.mp3",
                  "ML_prj_2023/D_5/I_0/Ashoor-Roohafza.mp3",
                  "ML_prj_2023/D_6/I_5/Siavash_kamkar_tarkibi_4gah.mp3"]
```

```python
def plot_mfcc(mfcc, sr, class_name):
  plt.figure(figsize=(10, 4))
  librosa.display.specshow(mfcc, sr=sr, x_axis='time')
  plt.colorbar()
  plt.title(f'MFCC (Class name: {class_name})')
  plt.tight_layout()
  plt.show()
  
```

```python
def plot_chroma(chroma, class_name):
  plt.figure(figsize=(10, 5))
  librosa.display.specshow(chroma, y_axis='chroma', x_axis='time')
  plt.colorbar()
  plt.title(f'Chroma feature (Class name: {class_name})')
  plt.tight_layout()
  plt.show()
```

```python
for class_index, file_path in enumerate(selected_files):
  class_name = class_names[class_index]
  y , sr = librosa.load(file_path)

  mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20,n_fft = 2048, hop_length = 512)
  chroma = librosa.feature.chroma_stft(y=y, sr=sr, n_chroma=12 ,n_fft = 2048, hop_length = 512)

  print(f"file_path: {file_path}")
  plot_mfcc(mfcc, sr, class_name)
  plot_chroma(chroma, class_name)
  print()
```
